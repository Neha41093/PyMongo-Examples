### Geospatial Dataset: Shipwrecks
*Given below is a small example on how to view geospatial data using the 'folium' package in Python*

```python
import pymongo
from pymongo import MongoClient
import folium
```

```python
client = MongoClient('localhost', 27017)
db = client['ships']
collection = db['shipwrecks']

#Take a small sample to display correctly
ships_iterator = collection.find().limit(500)

mymap = folium.Map(location=[-76.87684485567021, 38.409596060278545 ], zoom_start=2)
i = 0
for ship in ships_iterator:
    if ship['coordinates']:
        folium.Marker(location=list(reversed(ship['coordinates']))).add_to(mymap)
    i +=1
    
mymap.save('map_shipwreclks.html')
display(mymap)
```
